## Import dataset

```python
azul_acoes = pd.read_csv('azul_acoes.csv')
corona_dolar = pd.read_csv('corona_dolar.csv')
```

## Cleaning dataset

```python
corona_dolar = corona_dolar.drop(columns=['Unnamed: 0'])
```

```python
corona_dolar = corona_dolar.rename(columns={'high': 'dollar_high', 'low': 'dollar_low'})
```

```python
corona_dolar
```

```python
azul_acoes = azul_acoes.rename(columns={'high': 'azul_high', 'low': 'azul_low'})
azul_acoes
```

## Merge two datasets

```python
df = pd.merge(left=corona_dolar, right=azul_acoes, how='left')
df.head(5)
```

## Export to CSV

```python
df.to_csv('data.csv')
```

## Interactive line chart

### Total New Cases

```python
df.iplot(kind='line', x='date', y='newcases', color='blue') 
```

### Azul

```python
fig = go.Figure(data=[go.Candlestick(x=df['date'],
                open=df['open'], high=df['azul_high'],
                low=df['azul_low'], close=df['close'],
                increasing_line_color= 'cyan', decreasing_line_color= 'gray')])
fig.show()
```
