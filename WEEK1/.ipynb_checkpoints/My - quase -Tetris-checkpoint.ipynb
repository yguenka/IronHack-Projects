```python
import pygame
import random

pygame.font.init()

largura_janela = 800
altura_janela = 650
largura_jogo = 300
altura_jogo = 600
tamanho_quadrado = 30
largura_x = (largura_janela - largura_jogo) // 2
altura_y = altura_janela - altura_jogo
dx = largura_x - largura_jogo + 100
dy = altura_y + altura_jogo/2 - 100

S = [
['.....', 
 '......', 
 '..00..', 
 '.00...', 
 '.....'], 
['.....',
 '..0..', 
 '..00.', 
 '...0.',
 '.....'],
['.....', 
 '.....', 
 '.00..', 
 '..00.',
 '.....'], 
['.....', 
 '..0..', 
 '.00..', 
 '.0...',
 '.....']] 

I  =  [
['..0..', 
 '..0..', 
 '..0..', 
 '..0..', 
 '.....'], 
['.....', 
 '0000.', 
 '.....', 
 '.....', 
 '.....']]

J = [
['.....', 
 '.0...', 
 '.000.', 
 '.....', 
 '.....'], 
['.....', 
 '..00.', 
 '..0..', 
 '..0..', 
 '.....'], 
['.....', 
 '.....', 
 '.000.', 
 '...0.', 
 '.....'], 
['.....', 
 '..0.', 
 '..0..', 
 '.00..', 
 '.....']]

L = [
['.....',
 '...0.',
 '.000.',
 '.....',
 '.....'],
['.....',
 '..0..',
 '..0..',
 '..00.',
 '.....'],
['.....',
 '.....',
 '.000.',
 '.0...',
 '.....'],
['.....',
 '.00..',
 '..0..',
 '..0..',
 '.....']]

T = [
['.....',
 '..0..',
 '.000.',
 '.....',
 '.....'],
['.....',
 '..0..',
 '..00.',
 '..0..',
 '.....'],
['.....',
 '.....',
 '.000.',
 '..0..',
 '.....'],
['.....',
 '..0..',
 '.00..',
 '..0..',
 '.....']]

X = [
['.....',
 '..0..',
 '.000.',
 '..0..',
 '.....']]

O = [
['.....',
 '.00..',
 '.00..',
 '.....',
 '.....']]
pecas = [S, I, J, L, T, X, O]
peca_cor = [(255, 0, 150), (180, 31, 222), (0, 0, 255), (25, 255, 255), (255, 255, 0), (0, 255, 255), (0, 28, 255)]
espaco = []


class My_Tetris(object):
    def __init__(self, x, y, peca):
        self.x = x
        self.y = y
        self.peca = peca
        self.cor = peca_cor[pecas.index(peca)]
        self.rotacao = 0


def grade_jogo(posicao_fixa={}):
    grade = [[(0, 0, 0) for _ in range(10)] for _ in range(20)]
    for i in range(len(grade)):
        for j in range(len(grade[i])):
            if (j, i) in posicao_fixa:
                g = posicao_fixa[(j, i)]
                grade[i][j] = g
    return grade


def peca_rotacao(peca):
    posicoes = []
    formato_peca = peca.peca[peca.rotacao % len(peca.peca)]
    for i, linha in enumerate(formato_peca):
        linhas = list(linha)
        for j, coluna in enumerate(linhas):
            if coluna == '0':
                posicoes.append((peca.x + j, peca.y + i))
    return posicoes


def espaco_jogo(peca, grade):
    posicao_ok = [[(j, i) for j in range(10) if grade[i][j] == (0, 0, 0)] for i in range(20)]
    posicao_ok = [j for abc in posicao_ok for j in abc]
    formato = peca_rotacao(peca)
    for pos in formato:
        if pos not in posicao_ok:
            if pos[1] > -1:
                return False
    return True

def grade_tela(tela, grade):
    dx = largura_x
    dy = altura_y
    for i in range(len(grade)):
        pygame.draw.line(tela, (128, 128, 128), 
                         (dx, dy + i * tamanho_quadrado), (dx + largura_jogo, dy + i * tamanho_quadrado))
        for j in range(len(grade[1])):
            pygame.draw.line(tela, (128, 128, 128), (dx + j * tamanho_quadrado, dy),
                             (dx + j * tamanho_quadrado, dy + altura_jogo))

def perdeu(posicoes):
    for pos in posicoes:
        x, y = pos
        if y < 1:
            return True
    return False


def limpar_linhas(grade, travar):
    inicio = 0
    for i in range(len(grade)-1, -1, -1):
        linhas = grade[i]
        if (0, 0, 0) not in linhas:
            inicio += 1
            indice = i
            for j in range(len(linhas)):
                try:
                    del travar[(j, i)]
                except:
                    continue
    return inicio

def espaco_vago():
    global espaco
    if len(espaco) == 0:
        tem_espaco()
    bloco = random.choice(espaco)
    espaco.remove(bloco)
    return My_Tetris(5, 0, bloco)


def tem_espaco():
    global espaco
    espaco = pecas.copy()



def ver_proxima_peca(peca, tela):
    formato_peca = peca.peca[peca.rotacao % len(peca.peca)]
    for i, linha in enumerate(formato_peca):
        linhas = list(linha)
        for j, coluna in enumerate(linhas):
            if coluna == '0':
                pygame.draw.rect(tela, peca.cor, 
                                 (dx + j*tamanho_quadrado, dy + i*tamanho_quadrado, tamanho_quadrado, tamanho_quadrado), 0)

def textos_display(texto, tamanho, cor, tela):
    fonte = pygame.font.SysFont('arial', tamanho, bold=True)
    frase = fonte.render(texto, 1, cor)
    tela.blit(frase, 
              (largura_x + largura_jogo/2 - (frase.get_width()/2), altura_y + altura_jogo/2 - (frase.get_height()/2)))

    
def janela(tela, grade, pontos=0):
    tela.fill((0, 0, 0))
    for i in range(len(grade)):
        for j in range(len(grade[i])):
            pygame.draw.rect(tela, grade[i][j], 
                             (largura_x + j*tamanho_quadrado, altura_y +i*tamanho_quadrado, 
                              tamanho_quadrado, tamanho_quadrado), 0)
    pygame.draw.rect(tela, (128, 128, 128), (largura_x, altura_y, largura_jogo, altura_jogo), 5)
    grade_tela(tela, grade)


def vai():
    posicao_fixa = {}
    mudar_peca = False
    jogando = True
    peca_atual = espaco_vago()
    proxima_peca = espaco_vago()
    relogio = pygame.time.Clock()
    tempo = 0
    pontos = 0

    while jogando:
        grade = grade_jogo(posicao_fixa)
        velocidade = 0.40
        tempo += relogio.get_rawtime()
        relogio.tick()

        if tempo/800 > velocidade:
            tempo = 0
            peca_atual.y += 1
            if not espaco_jogo(peca_atual, grade) and peca_atual.y > 0:
                peca_atual.y -= 1
                mudar_peca = True

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                jogando = False

            if event.type == pygame.KEYDOWN:
                if event.key in (pygame.K_LEFT, pygame.K_a):
                    peca_atual.x -= 1
                    if not espaco_jogo(peca_atual, grade):
                        peca_atual.x += 1

                if event.key in (pygame.K_RIGHT, pygame.K_d):
                    peca_atual.x += 1
                    if not espaco_jogo(peca_atual, grade):
                        peca_atual.x -= 1

                if event.key in (pygame.K_DOWN, pygame.K_s):
                    peca_atual.y += 1
                    if not espaco_jogo(peca_atual, grade):
                        peca_atual.y -= 1

                if event.key in (pygame.K_UP, pygame.K_w):
                    peca_atual.rotacao += 1
                    if not espaco_jogo(peca_atual, grade):
                        peca_atual.rotacao -= 1

                if event.key in (pygame.K_q, pygame.K_ESCAPE):
                    jogando = False

        peca_posicao = peca_rotacao(peca_atual)

        for i in range(len(peca_posicao)):
            x, y = peca_posicao[i]
            if y > -1:
                grade[y][x] = peca_atual.cor

        if mudar_peca:
            for pos in peca_posicao:
                p = (pos[0], pos[1])
                posicao_fixa[p] = peca_atual.cor

            peca_atual = proxima_peca
            proxima_peca = espaco_vago()
            mudar_peca = False
            pontos += limpar_linhas(grade, posicao_fixa)
    
        janela(ganhou, grade)
        ver_proxima_peca(proxima_peca, ganhou)
        pygame.display.update()

        if perdeu(posicao_fixa):
            textos_display("Você perdeu :/", 80, (255, 255, 255), ganhou)
            pygame.display.update()
            pygame.time.delay(1500)
            run = False


def foi():
    jogando = True
    while jogando:
        ganhou.fill((0, 0, 0))
        textos_display("Aperte qualquer tecla para começar !", 30, (255, 255, 255), ganhou)
        pygame.display.update()
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                jogando = False
            if event.type == pygame.KEYDOWN:
                vai()
    pygame.display.quit()


ganhou = pygame.display.set_mode((largura_janela, altura_janela))
pygame.display.set_caption('Teeeeetris')
foi()
```
