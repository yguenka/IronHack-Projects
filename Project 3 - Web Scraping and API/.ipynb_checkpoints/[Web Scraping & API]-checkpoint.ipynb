# Web Scraping

## Import libraries


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from datetime import date, datetime
```

## Select  website

```python
url = 'https://www.worldometers.info/coronavirus/country/brazil/'
```

## Extracting date

```python
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html)
data = str(soup.find_all('script'))
```

## Cleaning dataset


```python
new_cases_data = re.findall('data: .[\w{1,5},]{1,1000}.', data)[5].replace('data: [', '').replace('"', '').replace(']','').split(',')
```

```python
new_cases_days = re.findall('categories: .["\w{1,4} \d{1,3}.",]{1,1000}',data)[2].replace('categories: [', '').replace('"', '').split(',')
new_cases_days_format =[]
for data in new_cases_days:
    if 'Feb ' in data:
        new_cases_days_format.append(data.replace('Feb ', '2020-02-'))
    else: 
        new_cases_days_format.append(data.replace('Mar ', '2020-03-'))
```

```python
obj = {'date': new_cases_days_format, 'newcases': new_cases_data}
new_cases = pd.DataFrame(data=obj)
new_cases.head(5)
```

# API

## Connecting API and Extracting date - API de Cotações de Moedas

```python
import datetime
```

```python
response = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/41')
obj = response.json()
df = pd.DataFrame(data=obj).drop(columns=['code', 'codein', 'name', 'ask', 'pctChange', 'create_date'])
df.head(5)
```

## Cleaning dataset

```python
df['Data e Hora'] = pd.to_datetime(df['timestamp'], unit='s')
df['date'] = df['Data e Hora'].dt.strftime('%Y-%m-%d')
dolar = df.drop(columns=['timestamp', 'Data e Hora'])
dolar = dolar.drop([df.index[0], df.index[4], df.index[10], df.index[16],df.index[22], df.index[28], df.index[34], df.index[40]]).sort_values('date')
```

```python
dolar.head(5)
```

```python
result = pd.merge(left = new_cases, right = dolar, how='left')
```

## Export to CSV

```python
result.to_csv('corona_dolar.csv')
```
